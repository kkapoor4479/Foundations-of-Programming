# Task	1	-	Data	Analysis	with	Lists

A list of 10 random numbers representing daily sales figures is created and stored in a Python list. The program iterates through the list to print each value, then uses built-in functions sum(), len(), max(), and min() to compute the total, average, highest, and lowest sales. 

```python
# Task 1 — Data Analysis with Lists

from random import uniform

# Generate 10 random sales records between $1000 and $5000
daily_sales = [round(uniform(1000, 5000), 2) for i in range(10)]

# Calculate total, average, highest, and lowest values
total   = sum(daily_sales)
average = total / len(daily_sales)
highest = max(daily_sales)
lowest  = min(daily_sales)

# Print all the values
print("=" * 40)
print("          DAILY SALES REPORT")
print("=" * 40)

print("\nAll Sales Values:")
for i, sale in enumerate(daily_sales, start=1):
    print(f"  Day {i}: ${sale:,.2f}")

print("\n" + "-" * 40)
print(f"  Total Sales   : ${total:,.2f}")
print(f"  Average Sales : ${average:,.2f}")
print(f"  Highest Sales : ${highest:,.2f}")
print(f"  Lowest Sales  : ${lowest:,.2f}")
print("=" * 40)
```

# Task 2 — Conditional Logic: Grade Evaluation

This program accepts a numeric score from the user and uses a series of if-elif-else conditional statements to map the score to a letter grade (A to F). Input validation ensures the score stays within the valid 0–100 range, raising a clear error for out-of-range values. The program also determines pass/fail status by checking whether the score is 60 or above.

```python
# Task 2 — Conditional Logic: Grade Evaluation

score = float(input("Please enter the score"))

# Ensure score is valid (between 0 and 100)
if not (0 <= score <= 100):
    raise ValueError("Score must be between 0 and 100")

# Calculate grade
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

# Determine Pass or Fail
status = "Pass" if score >= 60 else "Fail"

# Print values
print("=" * 40)
print("          GRADE EVALUATION REPORT")
print("=" * 40)
print(f"\n  Score  : {score}")
print(f"  Grade  : {grade}")
print(f"  Status : {status}")
print("-" * 40)
```

# Task 3 — Function: Statistical Calculator

A reusable function calculate_statistics(numbers) is defined to accept any list of numbers and return three key statistics: the minimum, maximum, and average values. Python's built-in min(), max(), sum(), and len() are used. The function returns a dictionary so the caller can conveniently access each statistic by name.

```python
# Task 3 — Function: Statistical Calculator

from random import uniform

def calculate_statistics(numbers):
    # Check if list is empty
    if not numbers:
        raise ValueError("The list must not be empty.")

    # return min, max, and avg in a dict for easier use
    return {
        'minimum': min(numbers),
        'maximum': max(numbers),
        'average': sum(numbers) / len(numbers),
    }

# Generate 10 random numbers between 10 and 50
dataset = [round(uniform(10, 50), 2) for _ in range(5)]

stats = calculate_statistics(dataset)

# Print all the values
print("=" * 40)
print("          STATISTICAL CALCULATOR")
print("=" * 40)
print(f"\n  Dataset : {[data for data in dataset]}")
print(f"  Minimum : {stats['minimum']:.2f}")
print(f"  Maximum : {stats['maximum']:.2f}")
print(f"  Average : {stats['average']:.2f}")
print("-" * 40)
```

# Task 4 — Class Design: Student Record

A Student class is defined with three attributes — name, student_id, and grade; set through the `__init__` constructor. The display_info() method prints a summary of the student's record, and is_passed() returns a Boolean indicating whether the student's grade meets the passing threshold of 60. Three student objects are instantiated with different grades to demonstrate both passing and failing cases.

```python
# Task 4 — Class Design: Student Record

class Student:
    def __init__(self, name, student_id, grade):
        self.name       = name
        self.student_id = student_id
        self.grade      = grade

    def display_info(self):
        # Determine Pass or Fail based on the grade
        status = "Pass" if self.is_passed() else "Fail"

        # Print student details
        print(f"  Name       : {self.name}")
        print(f"  Student ID : {self.student_id}")
        print(f"  Grade      : {self.grade}")
        print(f"  Status     : {status}")
        print("-" * 40)

    def is_passed(self):
        # Determine Pass or Fail based on the grade
        return self.grade >= 60


students = [
    Student("Student A", "1001", 88),
    Student("Student B", "1002", 54),
    Student("Student C", "1003", 72),
]

print("=" * 40)
print("       STUDENT RECORD SYSTEM")
print("=" * 40)

for student in students:
    student.display_info()
```

# Task 5 — Mini Program: Bank Account Simulation

A BankAccount class is designed with account_holder and balance attributes initialised via __init__. The deposit() method adds a valid positive amount to the balance, while withdraw() checks for sufficient funds before subtracting the requested amount. display_balance() prints the current balance in a formatted currency style after each transaction.

```python
# Task 5 — Mini Program: Bank Account Simulation

class BankAccount:
    def __init__(self, account_holder, balance=0.0):
        self.account_holder = account_holder
        self.balance        = balance
        print(f"  Account created for: {self.account_holder}")
        print(f"  Opening Balance    : ${self.balance:,.2f}")
        print("-" * 40)

    def deposit(self, amount):
        # Ensure deposit amount > 0
        if amount <= 0:
            print(f"Deposit failed: amount must be positive (got ${amount:,.2f})")
            print("-" * 40)
            return
        self.balance += amount
        print(f"  Deposited : +${amount:,.2f}")
        self.display_balance()

    def withdraw(self, amount):
        # Ensure withdrawl amount > 0
        if amount <= 0:
            print(f"Withdrawal failed: amount must be positive (got ${amount:,.2f})")
            print("-" * 40)
            return

        # Ensure withdrawl amount < total balance
        if amount > self.balance:
            print(f"Withdrawal failed: insufficient funds "
                  f"(requested ${amount:,.2f}, available ${self.balance:,.2f})")
            print("-" * 40)
            return
        self.balance -= amount
        print(f"  Withdrew  : -${amount:,.2f}")
        self.display_balance()

    def display_balance(self):
        print(f"  Balance   :  ${self.balance:,.2f}")
        print("-" * 40)


print("=" * 40)
print("          BANK ACCOUNT SIMULATION")
print("=" * 40)
print()

account = BankAccount("KK", balance=500.00)

account.deposit(1500.00)
account.deposit(250.00)
account.withdraw(300.00)
account.withdraw(2500.00)   
account.deposit(-100)     
account.withdraw(400.00)

print()
print("=" * 40)
print(f"  Final Balance for {account.account_holder}: ${account.balance:,.2f}")
print("=" * 40)
```
